
### 3. Project Task: NMF recommender function

    Implement a recommender function that recommends movies to a new user based on the NMF model!



```python
def recommend_movies_to_new_user(sorted_list, model, k=10):
    """
    Recommends the top k movies for a new user based on an NMF model. 
    Returns a list of k movie ids.
    """
    
    recommendations = []

    # Create new user-item dataframe
    user_row = np.zeros((1, model.components_.shape[0]))
    for movie, rating in sorted_list.items():
        if movie in model.movies:
            # If the movie is in the model's movie list, add the rating to the appropriate column
            movie_idx = model.movies.index(movie)
            user_row[0, movie_idx] = rating

    # Calculate scores for each movie in the new user-item dataframe
    scores = np.dot(user_row, model.components_.T)[0]

    # Sort movies by score in descending order
    rated_movies = np.argsort(scores)[::-1]

    # Add the top-k highest rated movie ids to the recommendations list
    for movie_id in rated_movies:
        if model.movies[movie_id] not in sorted_list.keys():
            recommendations.append(model.movies[movie_id])
            if len(recommendations) == k:
                break

    return recommendations
```
